```python
class Rectangle:
    def __init__(self,x=0,y=0,w=0,h=0,r=120,g=20,b=220):
        self.x = x # Position X
        self.y = y # Position Y
        self.w = w # Width
        self.h = h # Height
        self.r = r # red
        self.g = g # green
        self.b = b # blue
    def draw(self,screen):
        pg.draw.rect(screen,(self.r,self.g,self.b),(self.x,self.y,self.w,self.h))
```

```python
class Button(Rectangle):
    def __init__(self, x=0, y=0, w=0, h=0, r=120, g=20, b=220):
        Rectangle.__init__(self, x, y, w, h, r, g, b)
        self.prev_mouse_pressed = False

    def isMouseOn(self):
        self.xn, self.yn = pg.mouse.get_pos()
        if (self.x <= self.xn <= self.x + self.w) and (self.y <= self.yn <= self.y + self.h):
            return True
        else:
            return False

    def isMousePressed(self):
        if self.isMouseOn():
            self.mouse_presses = pg.mouse.get_pressed()
            if self.mouse_presses[0] and not self.prev_mouse_pressed:
                print("Left Mouse key was clicked on the button")
            self.prev_mouse_pressed = self.mouse_presses[0]
        else:
            self.prev_mouse_pressed = False


```

```python
import sys 
import pygame as pg

pg.init()
run = True
win_x, win_y = 800, 480
screen = pg.display.set_mode((win_x, win_y))
btn = Button(20,20,100,100) # สร้าง Object จากคลาส Button ขึ้นมา
xa = False
xm = False
ya = False
ym = False
while run:
    screen.fill((255, 255, 255))
    

    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            run = False
        if event.type == pg.KEYDOWN and event.key == pg.K_w: 
            print("Key W down")
            ym = True
        if event.type == pg.KEYDOWN and event.key == pg.K_a: 
            print("Key A down")
            xm = True
        if event.type == pg.KEYDOWN and event.key == pg.K_s: 
            print("Key S down")
            ya = True
        if event.type == pg.KEYDOWN and event.key == pg.K_d: 
            print("Key D down")
            xa = True

        if event.type == pg.KEYUP and event.key == pg.K_w: 
            print("Key W up")
            ym = False
        if event.type == pg.KEYUP and event.key == pg.K_a: 
            print("Key A up")
            xm = False
        if event.type == pg.KEYUP and event.key == pg.K_s: 
            print("Key S up")
            ya = False
        if event.type == pg.KEYUP and event.key == pg.K_d: 
            print("Key D up")
            xa = False
    if xa:
        btn.x +=1
    if xm:
        btn.x -=1
    if ya:
        btn.y +=1
    if ym:
        btn.y -=1
    

    btn.draw(screen)

    pg.display.update()
```
