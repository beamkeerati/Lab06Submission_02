```python
import sys
import pygame as pg
class Rectangle:
    def __init__(self,x=0,y=0,w=0,h=0,r=120,g=20,b=220):
        self.x = x # Position X
        self.y = y # Position Y
        self.w = w # Width
        self.h = h # Height
        self.r = r # red
        self.g = g # green
        self.b = b # blue
    def draw(self,screen):
        pg.draw.rect(screen,(self.r,self.g,self.b),(self.x,self.y,self.w,self.h))
```

```python
class Button(Rectangle):
    def __init__(self, x=0, y=0, w=0, h=0, r=0, g=0, b=128):
        Rectangle.__init__(self, x, y, w, h, r, g, b)
        self.prev_mouse_pressed = False
        self.clicked = False

    def isMouseOn(self):
        self.xn, self.yn = pg.mouse.get_pos()
        if (self.x <= self.xn <= self.x + self.w) and (self.y <= self.yn <= self.y + self.h):
            return True
        else:
            return False

    def isMousePressed(self):
        self.clicked = False
        if self.isMouseOn():
            self.mouse_presses = pg.mouse.get_pressed()
            if self.mouse_presses[0] and not self.prev_mouse_pressed:
                #print("Left Mouse key was clicked on the button")
                self.clicked = True
            self.prev_mouse_pressed = self.mouse_presses[0]
        else:
            self.prev_mouse_pressed = False


```

```python
class Text:
    def __init__(self, text, x, y, font, color):
        self.text = text
        self.x = x
        self.y = y
        self.font = font
        self.color = color

    def draw(self, screen):
        text_surface = self.font.render(self.text, True, self.color)
        screen.blit(text_surface, (self.x, self.y))
```

```python
class InputBox:

    def __init__(self, x, y, w, h, text=''):
        self.rect = pg.Rect(x, y, w, h) 
        self.color = COLOR_INACTIVE
        self.text = text
        self.txt_surface = FONT.render(text, True, self.color)
        self.active = False

    def handle_event(self, event):
        if event.type == pg.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                self.active = not self.active
            else:
                self.active = False
            self.color = COLOR_ACTIVE if self.active else COLOR_INACTIVE
        if event.type == pg.KEYDOWN:
            if self.active:
                if event.key == pg.K_RETURN:
                    print(self.text)
                    self.text = ''
                elif event.key == pg.K_BACKSPACE:
                    self.text = self.text[:-1]
                else:
                    self.text += event.unicode
                self.txt_surface = FONT.render(self.text, True, self.color)

    def update(self):
        # Resize the box if the text is too long.
        width = max(200, self.txt_surface.get_width()+10)
        self.rect.w = width


    def  draw(self, screen):
        # Blit the text.
        screen.blit(self.txt_surface, (self.rect.x + 5, self.rect.y + 5))
        # Blit the rect.
        pg.draw.rect(screen, self.color, self.rect, 2)  # Fixed the method call
```

```python
import sys
import pygame as pg

pg.init()
win_x, win_y = 800, 480
screen = pg.display.set_mode((win_x, win_y))

# Load the icon image
icon = pg.image.load("icon.png")

# Scale the icon image
icon_width, icon_height = 64, 83  # Set the desired icon size
scaled_icon = pg.transform.scale(icon, (icon_width, icon_height))

COLOR_INACTIVE = pg.Color('lightskyblue3') # ตั้งตัวแปรให้เก็บค่าสี เพื่อนำไปใช้เติมสีให้กับกล่องข้อความตอนที่คลิกที่กล่องนั้นๆอยู่
COLOR_ACTIVE = pg.Color('dodgerblue2')     # ^^^
FONT = pg.font.Font("NotoSansThai-Regular.ttf", 24)

input_box1 = InputBox(100, 100, 140, 42) # สร้าง InputBox1
input_box2 = InputBox(100, 200, 140, 42) # สร้าง InputBox2
input_box3 = InputBox(100, 300, 140, 42) # 
input_boxes = [input_box1, input_box2,input_box3] # เก็บ InputBox ไว้ใน list เพื่อที่จะสามารถนำไปเรียกใช้ได้ง่าย

headline_font = pg.font.Font(None, 24)  
headline_color = pg.Color('black')      
headline1 = Text("Name", 100, 70, headline_font, headline_color)
headline2 = Text("Surname", 100, 170, headline_font, headline_color)
headline3 = Text("Age", 100, 270, headline_font, headline_color)
headline4 = Text("Submit", 450+35, 208, headline_font, pg.Color('white'))
headlines = [headline1, headline2,headline3,headline4]

btn = Button(450,200,140,32) # สร้าง Object จากคลาส Button ขึ้นมา


run = True

while run:
    screen.fill((255, 255, 255))

    icon_x = 5    # Set the x position 5 pixels from the right border
    icon_y = 5  # Set the y position 5 pixels from the top border
    screen.blit(scaled_icon, (icon_x, icon_y))
    
    btn.isMousePressed()
    if btn.isMouseOn():
        if btn.clicked:
            btn.r = 0
            btn.g = 128
            btn.b = 0
            
        else:
            btn.r = 0
            btn.g = 0
            btn.b = 128

    btn.draw(screen)

    for headline in headlines:
        headline.draw(screen)

    for box in input_boxes: # ทำการเรียก InputBox ทุกๆตัว โดยการ Loop เข้าไปยัง list ที่เราเก็บค่า InputBox ไว้
        box.update() # เรียกใช้ฟังก์ชัน update() ของ InputBox
        box.draw(screen) # เรียกใช้ฟังก์ชัน draw() ของ InputBox เพื่อทำการสร้างรูปบน Screen
        
    for event in pg.event.get():

        if event.type == pg.QUIT:
            pg.quit()
            run = False

        if event.type == pg.MOUSEBUTTONDOWN:
            btn.isMousePressed()
            if btn.clicked :
                if input_box1.text!="" and input_box2.text!="" and input_box3.text.isdigit() and 0<int(input_box3.text)<=100:
                    print("Name:", input_box1.text)
                    print("Surname:", input_box2.text)
                    print("Age:", input_box3.text)
                    screen.fill((255, 255, 255))
                    pg.display.update()
                    run = False
                    break
                else:
                    Noti = Text(f"Provide full name and numeric age; noncompliance may cause issues.", 100,400, pg.font.Font(None, 24), pg.Color('red'))
                    Noti.draw(screen)
                    pg.display.update()
                    pg.time.delay(2000)

        for box in input_boxes:
            box.handle_event(event)

    pg.time.delay(1)
    pg.display.update()

headline6 = Text(f"Hello {input_box1.text} {input_box2.text}! You are {input_box3.text} years old.", 80, 240, pg.font.Font("NotoSansThai-Regular.ttf", 36), headline_color)

run = True

while run:
    screen.fill((255, 255, 255))
    headline6.draw(screen)
    icon_x = (win_x - icon_width) // 2    # Set the x position 5 pixels from the right border
    icon_y = 20  # Set the y position 5 pixels from the top border
    screen.blit(scaled_icon, (icon_x, icon_y))
    for event in pg.event.get():

        if event.type == pg.QUIT:
            pg.quit()
            run = False

    pg.time.delay(1)
    pg.display.update()
```
