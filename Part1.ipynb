```python
class Rectangle:
    def __init__(self,x=0,y=0,w=0,h=0,r=120,g=20,b=220):
        self.x = x # Position X
        self.y = y # Position Y
        self.w = w # Width
        self.h = h # Height
        self.r = r # red
        self.g = g # green
        self.b = b # blue
    def draw(self,screen):
        pg.draw.rect(screen,(self.r,self.g,self.b),(self.x,self.y,self.w,self.h))
```

```python
import sys 
import pygame as pg

pg.init()
run = True
win_x, win_y = 800, 480
screen = pg.display.set_mode((win_x, win_y))
firstObject = Rectangle(20,20,100,100) # สร้าง Object จากคลาส Rectangle ขึ้นมา

while(run):
    screen.fill((255, 255, 255))
    firstObject.draw(screen) # ใส่ screen เข้าไปด้วยเพราะว่าคำสั่ง pg.draw.rect จะเป็นจะต้องระบุระนาบว่าต้องการสร้างรูปบนระนาบใด
    pg.display.update()
    
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            run = False
```

```python
class Button(Rectangle):
    def __init__(self, x=0, y=0, w=0, h=0, r=120, g=20, b=220):
        Rectangle.__init__(self, x, y, w, h, r, g, b)
        self.prev_mouse_pressed = False

    def isMouseOn(self):
        self.xn, self.yn = pg.mouse.get_pos()
        if (self.x <= self.xn <= self.x + self.w) and (self.y <= self.yn <= self.y + self.h):
            return True
        else:
            return False

    def isMousePressed(self):
        if self.isMouseOn():
            self.mouse_presses = pg.mouse.get_pressed()
            if self.mouse_presses[0] and not self.prev_mouse_pressed:
                print("Left Mouse key was clicked on the button")
            self.prev_mouse_pressed = self.mouse_presses[0]
        else:
            self.prev_mouse_pressed = False
```

```python
import sys 
import pygame as pg

pg.init()
run = True
win_x, win_y = 800, 480
screen = pg.display.set_mode((win_x, win_y))
btn = Button(20,20,100,100) # สร้าง Object จากคลาส Button ขึ้นมา


while run:
    screen.fill((255, 255, 255))
    
    btn.isMousePressed()
    if btn.isMouseOn():
        btn.w = 200
        btn.h = 300
        if btn.prev_mouse_pressed:
            btn.r = 159
            btn.g = 43
            btn.b = 104
        else:
            btn.r = 220
            btn.g = 220
            btn.b = 220
    else:
        btn.w = 100
        btn.h = 100
        btn.r = 136
        btn.g = 8
        btn.b = 8
        
    btn.draw(screen)

    pg.display.update()
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            run = False
```

```python
import sys 
import pygame as pg

pg.init()
run = True
win_x, win_y = 800, 480
screen = pg.display.set_mode((win_x, win_y))
btn = Button(20,20,100,100) # สร้าง Object จากคลาส Button ขึ้นมา

while(run):
    screen.fill((255, 255, 255))
    if btn.isMouseOn():
        btn.w = 200
        btn.h = 300
    else:
        btn.w = 100
        btn.h = 100
    btn.draw(screen)
    
    pg.display.update()
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            run = False
```
